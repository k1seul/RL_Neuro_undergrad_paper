```python
from matplotlib import pyplot as plt
import math
import numpy as np
```

```python
negative_p = 0.5
log_weights = np.array([2, 4, 6, 8])
log_bias = log_weights * math.log(negative_p + 0.1)
```

```python
dx = np.arange(0, 1, 0.01)
plt.figure(figsize=(10,8))
for i in range(len(log_weights)):
    dy = -log_weights[i] * np.log(dx + 0.1) + log_bias[i]
    plt.plot(dx,dy,label=f"log weight = {log_weights[i]}", linewidth=2)
    
plt.axhline(y=0, color='k', linewidth=2)
plt.axvline(x=0.1, color='b', linewidth=2, label="p = 0.1")
plt.xlim([0,1])
plt.yticks(np.arange(-5, 4, 1))
plt.xticks(np.arange(0, 1, 0.1))
plt.xlabel("probability", fontdict = {'fontsize' : 12})
plt.ylabel("internal reward", fontdict = {'fontsize' : 12})
plt.title("Internal Reward Function", fontdict = {'fontsize' : 15})
plt.legend()
    
```

```python
sigmoid_weights = [2, 3, 4, 5, 10, 20] 

dx = np.arange(0, 1, 0.01)
plt.figure(figsize=(10,8))
for i in range(len(sigmoid_weights)):
    dy = 1 -  1 / (1 + ((dx+0.01)/(1-dx))**(-sigmoid_weights[i]))
    plt.plot(dx,dy,label=f"sigmoid weight = {sigmoid_weights[i]}", linewidth=2)

plt.axhline(y=0, color='k', linewidth=2)
plt.axvline(x=0.1, color='b', linewidth=2, label="p = 0.1")
plt.xlim([0,1])
plt.yticks(np.arange(-5, 4, 1))
plt.xticks(np.arange(0, 1, 0.1))
plt.xlabel("probability", fontdict = {'fontsize' : 12})
plt.ylabel("internal reward", fontdict = {'fontsize' : 12})
plt.title("Internal Reward Function", fontdict = {'fontsize' : 15})
plt.legend()
    
```
